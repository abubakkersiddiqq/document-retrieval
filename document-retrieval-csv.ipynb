```python
import chromadb
from chromadb.utils import embedding_functions


client = chromadb.Client()
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
        model_name="all-mpnet-base-v2" # You can choose any available Sentence Transformer model
    )
```

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/food-sample/FOOD-DATA-GROUP1.csv")

df.drop(['Unnamed: 0.1','Unnamed: 0'], axis=1, inplace=True)

df.columns = df.columns.str.lower()

columns = df.columns.tolist()

ids = [str(i) for i in range(1, len(df) + 1)]

documents = []

for i, row in df.iterrows():
    doc = " ".join([f"{col}: {value}" for col, value in row.items()])
    documents.append(doc)

metadata_list = df.to_dict(orient="records")
```

```python
def extract_column_fromquery(query,columns):
    query = query.lower()
    
    for col in columns:
        if col in query:
            return col
    return None
```

```python
#s1 delete collection name

#s2 manually to do embeddings doc + query
user_input = input("Enter your question: ").lower()

collection  = client.get_or_create_collection(
    name="my_collection",
    embedding_function =sentence_transformer_ef

)
collection.upsert(
        documents= documents,
        ids=ids,
        metadatas = metadata_list
    
    )
n = 3
result = collection.query(
    query_texts =[user_input],
    n_results  = n
)

col = extract_column_fromquery(user_input,columns)
print(f"The top {n} results:")
for i in range(n):
    row = result["metadatas"][0][i]
    print(row[col])
```
