```python
import chromadb
from chromadb.utils import embedding_functions


client = chromadb.Client()
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
        model_name="all-mpnet-base-v2" # u can choose any available Sentence Transformer model
    )
```

```python
import pandas as pd
from rapidfuzz import process , fuzz
```

```python
df = pd.read_csv("/kaggle/input/food-sample/FOOD-DATA-GROUP1.csv")[0:20]

df.drop(['Unnamed: 0.1','Unnamed: 0'], axis=1, inplace=True)

df.columns = df.columns.str.lower()

columns = df.columns.tolist()

ids = [str(i) for i in range(1, len(df) + 1)]

documents = []

for i, row in df.iterrows():
    doc = " ".join([f"{col}: {value}" for col, value in row.items()])
    documents.append(doc)

metadata_list = df.to_dict(orient="records")
```

```python
def detect_column_from_query(query, columns):
    query_words = query.lower().split()

    best_col = None
    best_score = 0

    for word in query_words:
        match, score, _ = process.extractOne(
            word,
            columns,
            scorer=fuzz.WRatio
        )
        if score > best_score:
            best_col = match
            best_score = score

    return best_col if best_score >= 65 else None
```

```python
#s1 delete collection name

#s2 manually to do embeddings doc + query
user_input = input("Enter your question: ").lower()

collection  = client.get_or_create_collection(
    name="my_collection",
    embedding_function =sentence_transformer_ef

)
collection.upsert(
        documents= documents,
        ids=ids,
        metadatas = metadata_list
    
    )
n = 3
result = collection.query(
    query_texts =[user_input],
    n_results  = n
)

col = detect_column_from_query(user_input, columns)
if col is None:
    print("\nCould not detect any column from your question.")
    print("Try again with clearer wording.\n")
else:
    print(f"\n Detected column: {col}\n")
    print(f"Top {n} answers:")

    for i in range(n):
        row_meta = result["metadatas"][0][i]
        print(f"- {row_meta[col]}")
```
